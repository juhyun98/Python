```python
import matplotlib.pyplot as plt
```

```python
y = [2, 4, 6]

plt.plot(y)
plt.show()
```

```python
y = [2, 4, 6]
x = [1, 2, 3]

# plot - 선 그래프(Line plot)

plt.plot(x, y)
# x, y의 갯수가 맞지 않으면 에러! -> 꼭 맞춰주자!
plt.show()
```

```python
# line style -> ls
plt.plot(x, y, ls = '--')
plt.show()
```

```python
# Marker
plt.plot(x, y, marker='o')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10, 13, 16, 20]
y2 = [10, 20, 15, 12]

plt.plot(x, y, ls = '-', marker = 'p')    # 그래프 그리기
plt.plot(x, y2, ls = ':')
plt.show()                  # 그래프 보여주기
```

```python
# line width 선 두께 -> lw
plt.plot(x, y, lw = 3)
plt.show()
```

```python
# line color 선 색상 변경
# RGB 색상 적용
plt.plot(x, y, color = '#ACFADF', marker = 'D')
plt.show()
```

```python
# marker color 마커 색상 - markerfacecolor -> mfc
plt.plot(x, y, color='#ACFADF', marker='D', mfc='#FBA1B7')
plt.show()
```

```python
# 마커 선 색상 - markeredgecolor -> mec
plt.plot(x, y, color='#ACFADF', marker='D', mfc='#FBA1B7', mec='#6528F7')
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw = 5, color='purple', marker='o', ms='15', mec='green', mew='5', mfc='red')
plt.show()
```

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw = 5, color='purple', marker='o', ms='15', mec='green', mew='5', mfc='red')
plt.xlim(-2, 8)        # x축 범위 지정    (최소, 최대)
plt.ylim(-1, 13)       # y축 범위 지정
plt.show()
```

```python
# 틱의 위치를 보기 위해. 격자무늬 표현
# tick - 그래프 축의 위치
plt.plot(x, y, ls='--', lw = 5, color='purple', marker='o', ms='15', mec='green', mew='5', mfc='red')
plt.xticks( [0, 3, 6] )
plt.yticks([1, 2, 3])
plt.grid()
plt.show()
```

```python
# 여러 plot 그리기
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [3, 6, 9, 12]

# 그래프에 이름 부여
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend()
plt.show()
```

```python
plt.plot(x, label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
plt.legend()
plt.show()
```

```python
# 한글 지원 폰트 변경
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
plt.plot(x, label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
# 축에 이름 표시
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)
# 그래프의 이름
plt.title('x, y, z 그래프')
plt.legend()
plt.show()
```
