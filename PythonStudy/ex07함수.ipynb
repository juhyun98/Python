#### 함수


```python
def number_sum(num1, num2) :
    r = num1 + num2
    return r
```

```python
print(number_sum(2, 3))
```

```python
def cal(num1, num2 = 1, op = "+") : 
    "덧셈과 뺄셈의 결과"
    if op == '+' :
        return num1 + num2
    elif op == '-' :
        return num1 - num2
    else :
        return '연산기호 입력 에러'
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(+, -) >> ")
result = cal(num1, num2, op)
print("결과 : {}".format(result))
```

```python
print(cal(7, op='-'))
print(cal(8))
```

```python
def add(*args) :
    sum = 0
    for i in args :
        sum += i
    return(sum)
```

```python
print(add(1,2,3))
print(add(4,5,6,7,8,9,10))
```

```python
# return 키워드는 한번만 사용 가능하다.

def add_sub(n1, n2) :
    return n1 + n2, n1 - n2, "더하기 빼기 결과"
```

```python
ar, sr, str1 = add_sub(3, 9)

print(ar)
print(sr)
print(str1)
```
