### 실습문제


```python
score = int(input("점수 입력 >> "))
if score >= 90 :
    print(score, "점은 A학점 입니다.")
elif score >= 80 :
    print(score, "점은 B학점 입니다.")
elif score >= 70 :
    print(score, "점은 C학점 입니다.")
elif score >= 60 :
    print(score, "점은 D학점 입니다.")
else :
    print(score, "점은 F학점 입니다.")
```
